```python
# importing necessary libraries
```

```python
import pandas as pd
import numpy as np
```

```python
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
%matplotlib inline
```

```python
df = pd.read_csv(r"C:\Users\NNAMDI\Desktop\CC GENERAL.csv")
df
```

```python
# removed the ID
```

```python
X = df[df.columns[1:18]]
X
```

```python
X.describe()
```

```python
X.isna().sum()
```

```python
# dropped null cells in MINIMUM_PAYMENTS columns
```

```python
X = X.dropna(subset = ["MINIMUM_PAYMENTS"])
```

```python
X.isna().sum()
```

```python
# imputing null value in CREDIT_LIMIT with the mean of the column
```

```python
from sklearn.impute import SimpleImputer
```

```python
imputer = SimpleImputer(strategy = 'mean')
```

```python
X_imputed = imputer.fit_transform(X)
```

```python
original_column = X.columns
```

```python
imputed_df = pd.DataFrame(X_imputed, columns = original_column)
imputed_df
```

```python
imputed_df.isna().sum()
```

```python
# scaling the dataset
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaled_df = scaler.fit_transform(imputed_df)
scaled_df
```

```python
refined_df = pd.DataFrame(scaled_df, columns = original_column)
refined_df
```

```python
# performing clustering using DBSCAN
```

```python
from sklearn.cluster import DBSCAN
```

```python
db_model = DBSCAN(eps = 1.1, min_samples = 4)
```

```python
db_model.fit(refined_df)
```

```python
db_model.labels_
```

```python
# performing clustering using Kmeans
```

```python
from sklearn.cluster import KMeans
```

```python
# created an elbow curve to know the optimal number of clusters
```

```python
option = range(2,10)
inertias = []

for n_clusters in option:
    k_model = KMeans(n_clusters, random_state = 42).fit(refined_df)
    inertias.append(k_model.inertia_)

plt.plot(option, inertias, '-o')   
plt.xlabel('no. of clusters')
plt.ylabel('inertia')
plt.show() 
```

```python
k_model = KMeans(n_clusters = 3, random_state = 42)
```

```python
k_model.fit(refined_df)
```

```python
k_model.inertia_
```

```python
k_model.cluster_centers_
```

```python
preds = k_model.predict(refined_df)
preds
```

```python
# reducing the dimensions using PCA to visualize the clusters
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components = 3)
```

```python
transformed = pca.fit_transform(refined_df)
transformed
```

```python
df_columns = ["X1","X2","X3"]
```

```python
new_df = pd.DataFrame(transformed, columns = df_columns)
new_df
```

```python
# visualization of KMeans clustering
```

```python
fig = px.scatter_3d(new_df, x = "X1", y = "X2", z = "X3", color = preds)
center_x, center_y = k_model.cluster_centers_[:,0], k_model.cluster_centers_[:,1]
fig.update_traces(marker_size = 5)
fig.show()
```

```python
# visualization of DBSCAN clustering
```

```python
fig = px.scatter_3d(new_df, x = "X1", y = "X2", z = "X3", color = db_model.labels_)
fig.update_traces(marker_size = 5)
fig.show()
```
